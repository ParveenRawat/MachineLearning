```python
def remove_punc(text):
    trans = str.maketrans("","",string.punctuation)
    return text.translate(trans)
```

```python
data["sms"] = data["sms"].apply(remove_punc)
data
```

```python
def remove_noise(text):
    t = re.sub('[^a-zA-Z]',' ',text)
    return t
```

```python
data
```

```python
data["sms"] = data["sms"].apply(remove_noise)
data
```

```python
from nltk.corpus import stopwords
sw = stopwords.words("english")
```

```python
sw
```

```python
def remove_sws(text):
    s = [word.lower() for word in text.split() if word.lower not in sw]
    return " ".join(s)
```

```python
data["sms"] = data["sms"].apply(remove_sws)
```

```python
data
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemma(text):
    l = [lemmatizer.lemmatize(word) for word in text.split()]
    
    return " ".join(l)
```

```python
data["sms"] = data["sms"].apply(lemma)
```

```python
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(max_features=3000)
```

```python
X = tf.fit_transform(data["sms"]).toarray()
Y = data["label"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,f1_score
```

```python
clf = BernoulliNB()
clf
```

```python
clf.fit(X_train,Y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
print("Accuracy is : ",accuracy_score(Y_test,y_pred))
```

```python
print("F1Score is : ",f1_score(Y_test,y_pred))
```

```python
print("confusion matrix : ",confusion_matrix(Y_test,y_pred))
```

```python
cm = confusion_matrix(Y_test,y_pred)
plt.figure(figsize = (5,5))
sns.heatmap(cm,annot = True, linewidth = 0.5, cmap="tab20",fmt = "d")
plt.title("NBC")
plt.xlabel("Predicted")
plt.xlabel("Truth")
```
